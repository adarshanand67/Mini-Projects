```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

```

```python
df = pd.read_csv('algoCompare.csv')
df.head()
```

```python
df = df[df['Time/Node'] < 0.00025]
df.head()
```

```python
sns.scatterplot(y=df.TimeTaken,x=df.Nodes, hue = df.Algorithm)
plt.show()
```

```python
sns.lineplot(x=df.Nodes, y= df["TimeTaken"], hue = df.Algorithm )
plt.show()
```

```python
sns.violinplot(x=df.Algorithm, y=df.TimeTaken)
plt.show()
```
